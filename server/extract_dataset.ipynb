## Dataset webscraped from Wikipedia:

```python
def get_wikipedia_sentences(article_title):
    base_url = "https://en.wikipedia.org/wiki/"
    url = base_url + article_title.replace(" ", "_")

    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, "html.parser")

        #trying to avoid certain types of sentences here
        skip_sections = {"etymology", "origins", "history", "taxonomy"}
        paragraphs = []
        current_section = ""

        #avoiding those headers
        for tag in soup.find_all(["h2", "h3", "p"]):
            if tag.name in ["h2", "h3"]:
                current_section = tag.get_text().strip().lower()
            elif tag.name == "p" and not any(sec in current_section for sec in skip_sections):
                paragraphs.append(tag.get_text())

        sentences = []
        for p in paragraphs:
            text = p.strip()
            sentence_list = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)

            for sentence in sentence_list:
                #avoiding historial sentences
                if article_title.lower() in sentence.lower() and not re.search(r'\b(1[5-9]\d{2}|20\d{2}|century|medieval|previously|historical)\b', sentence):
                    sentences.append(sentence.strip())

        return sentences

    except requests.exceptions.RequestException as e:
        print(f"Error fetching Wikipedia article: {e}")
        return []

all_sentences = []

for article in word_list:
    all_sentences.extend(get_wikipedia_sentences(article))

wikipedia_webscraped_df = pd.DataFrame(all_sentences)
wikipedia_webscraped_df.head()
```

```python
from deep_translator import GoogleTranslator
from korean_romanizer.romanizer import Romanizer

translator = GoogleTranslator(source="en", target="ko")

translated_data = []

for word in word_list:
    korean_word = translator.translate(word)
    romanized = Romanizer(korean_word).romanize()
    translated_data.append([word, romanized])

korean_translated_df = pd.DataFrame(translated_data, columns=["English", "Korean"])
```
